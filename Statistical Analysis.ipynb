```python
# Import necessary libraries
import numpy as np
import statistics # We'll use this for the mode

# --- Section A: Question 1 ---

# Data: Daily patient admissions for the past 10 days
admissions = np.array([32, 28, 35, 30, 29, 27, 31, 34, 33, 30])
print(f"Data: {admissions}")

# ● Compute the mean, median, and mode of patient admissions.

# Calculate Mean using numpy
mean_admissions = np.mean(admissions)

# Calculate Median using numpy
median_admissions = np.median(admissions)

# Calculate Mode using statistics.mode
mode_admissions = statistics.mode(admissions)

print(f"\nCalculated Measures:")
print(f"Mean Admissions: {mean_admissions:.1f}")
print(f"Median Admissions: {median_admissions:.1f}")
print(f"Mode Admissions: {mode_admissions}")

# Determine best measure of central tendency
print("\nThe mean best represents patient admissions as the data is relatively symmetric without outliers.\n")

# Calculate effect of 10% capacity increase
increased_admissions = [x * 1.1 for x in admissions]
new_mean = statistics.mean(increased_admissions)
new_median = statistics.median(increased_admissions)
new_mode = statistics.mode(increased_admissions)  # Mode after increase is 33.0

print(f"After a 10% capacity increase:")
print(f"New Mean: {new_mean:.2f}")
print(f"New Median: {new_median:.2f}")
print(f"New Mode: {new_mode:.2f}")
```

```python

# Data: Recovery duration (in days) of 10 patients
recovery_days = np.array([5, 7, 6, 8, 9, 5, 6, 7, 8, 6])
print(f"Data: {recovery_days}")

# ● Calculate the range, variance, and standard deviation.

# Calculate Range using numpy's ptp (peak-to-peak) function
range_recovery = np.ptp(recovery_days)

# Calculate Sample Variance using numpy.var with ddof=1
# ddof=1 (Delta Degrees of Freedom) ensures we calculate the SAMPLE variance (dividing by N-1)
variance_recovery = np.var(recovery_days, ddof=1)

# Calculate Sample Standard Deviation using numpy.std with ddof=1
# ddof=1 ensures we calculate the SAMPLE standard deviation (sqrt of sample variance)
std_dev_recovery = np.std(recovery_days, ddof=1)

print(f"\nCalculated Measures:")
print(f"Range: {range_recovery} days")
print(f"Sample Variance: {variance_recovery:.3f} days²")
print(f"Sample Standard Deviation: {std_dev_recovery:.3f} days")

# ● What does the standard deviation indicate about variability in recovery times?
mean_recovery = np.mean(recovery_days) # Calculate mean for context
print(f"The standard deviation of approximately {std_dev_recovery:.2f} days measures the typical")
print(f"spread or dispersion of individual recovery times around the average recovery")
print(f"time ({mean_recovery:.1f} days) for this group of patients.")
print("A larger standard deviation would mean recovery times are more spread out;")
print("a smaller one means they are more clustered around the average.")

# ● If two new patients take 4 and 10 days to recover, how will this impact the standard deviation?
print("\nImpact of adding two new patients (4 and 10 days):")

# Original data and standard deviation (calculated above)
print(f"Original Data: {recovery_days}")
print(f"Original Sample Standard Deviation: {std_dev_recovery:.3f} days")

# Create the new data set including the two new patients
new_recovery_days = np.append(recovery_days, [4, 10])
print(f"New Data: {new_recovery_days}")

# Calculate the new sample standard deviation (ddof=1) for the updated data
new_std_dev_recovery = np.std(new_recovery_days, ddof=1)
new_mean_recovery = np.mean(new_recovery_days) # For context

print(f"New Sample Standard Deviation: {new_std_dev_recovery:.3f} days")
# print(f"(New Mean: {new_mean_recovery:.2f})") # Optional: show new mean

# Explain the impact
if new_std_dev_recovery > std_dev_recovery:
    print("\nThe standard deviation has increased.")
    print("Reason: The added recovery times (4 and 10 days) are further away from the mean")
    print(f"(original mean was {mean_recovery:.1f}, new mean is {new_mean_recovery:.2f}) than the typical spread")
    print("of the original data points. Adding these more extreme values increases the overall")
    print("variability (spread) of the dataset.")
elif new_std_dev_recovery < std_dev_recovery:
     print("\nThe standard deviation has decreased.") # This is not the case here
else:
     print("\nThe standard deviation remained the same.") # Very unlikely
```
